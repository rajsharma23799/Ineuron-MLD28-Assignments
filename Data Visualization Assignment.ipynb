```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris.keys()
```

```python
#for description of iris dataset
print(iris['DESCR'])
```

```python
iris['target_names']
```

```python
df=pd.DataFrame(iris['data'],columns=iris['feature_names'])
df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
t_data=pca.transform(scaled_data)
```

```python
scaled_data.shape  
```

```python
t_data.shape
```

```python
#transformed iris dataset
new_df=pd.DataFrame(t_data,columns=['vector1','vector2','vector3'])
new_df
```
