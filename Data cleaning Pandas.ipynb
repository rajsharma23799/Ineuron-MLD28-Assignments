```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
#Question 1
df['FlightNumber'][1]=10055
df['FlightNumber'][3]=10075
df['FlightNumber']=df['FlightNumber'].astype(int)
df
```

```python
#Question 2
df1=df.From_To.str.split('_',expand=True)
df1.columns=['From','To']
df1
```

```python
#Question 3
df1['From']=df1['From'].str.capitalize()
df1['To']=df1['To'].str.capitalize()
df1
```

```python
#Question 4
df=df.drop('From_To',axis=1)
df=df1.join(df)
df
```

```python
#Question 5
delays=pd.DataFrame(df['RecentDelays'].to_list(),columns=['delay_1','delay_2','delay_3'])
df=df.drop('RecentDelays',axis=1)
df=df.join(delays)
df=df[['From','To','FlightNumber','delay_1','delay_2','delay_3','Airline']]
df
```
